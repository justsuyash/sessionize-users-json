```python
import pandas as pd
import json 
import statistics
import urllib
import numpy as np 
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
import os
pd.set_option('display.max_columns', 10)
pd.set_option('display.max_rows',350)
```

```python
json_filepath = os.path.join('sessionize_data', 'city_searches.json')
```

```python
with open(json_filepath) as json_file:
    data_json = json.load(json_file)
```

```python
data_json[1]['user'][0][0]['user_id']
```

```python
## Lets flatten this and make this into a dataframe
## Note to self: Can be used as pointer in my unoversity lectures
```

```python
data = pd.DataFrame(columns = ["session_id", "unix_timestamp", "cities", "user_id","joining_date", "country"])

for i in range(len(data_json)):
    data.loc[i,'session_id']= data_json[i]['session_id'][0]
    data.loc[i,'unix_timestamp']= data_json[i]['unix_timestamp'][0]
    data.loc[i,'cities']= data_json[i]['cities'][0]
    data.loc[i,"user_id"] = data_json[i]['user'][0][0]['user_id']
    data.loc[i,"joining_date"] = data_json[i]['user'][0][0]['joining_date']
    data.loc[i,"country"] = data_json[i]['user'][0][0]['country']
```
